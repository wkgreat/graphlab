```python
import json
import matplotlib.pyplot as plt
with open("curvatures.json", "r", encoding="utf-8") as f:
    curvatures = json.load(f)
plt.figure(figsize=(50,10))
plt.xlim((0,1))
counts, bins, patches = plt.hist(curvatures,bins=10000)
patches
```
